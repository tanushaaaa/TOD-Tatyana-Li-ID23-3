3. По данным из файла `addres-book-q.xml` сформировать список словарей с телефонами каждого из людей. 

````python
from bs4 import BeautifulSoup  

with open("addres-book-q.xml", "r") as file:
    content = file.read()

soup = BeautifulSoup(content, "xml")

people_list = []

for address in soup.find_all("address"):
    person = {
        "name": address.find("name").text, 
        "phones": [phone.text for phone in address.find_all("phone")]  
    }
    people_list.append(person)

# Вывод списка людей с их телефонами
for person in people_list:
    print("Name:", person["name"])
    print("Phones:", person["phones"])
    print()
````

````python
def find_user_by_username(username):
    # Поиск пользователя по username
    for contributor in contributors_data:
        if contributor["username"] == username:
            # Вывод информации о пользователе
            print("Username:", contributor["username"])
            print("Name:", contributor["name"])
            print("Sex:", contributor["sex"])
            print("Address:", contributor["address"])
            print("Mail:", contributor["mail"])
            print("Jobs:", ", ".join(contributor["jobs"]))
            print("ID:", contributor["id"])
            return
    # Если пользователь не найден, возбуждение исключения
    raise ValueError(f"Пользователь с username '{username}' не найден.")
# Пример использования функции
try:
    find_user_by_username("uhebert")
except ValueError as e:
    print(e)
````

### XML

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python
with open('steps_sample.xml', 'r', encoding='utf-8') as file:
    xml_content = file.read()
# Создаем объект BeautifulSoup для парсинга XML
soup = BeautifulSoup(xml_content, 'xml')
# Инициализируем словарь для хранения шагов по каждому рецепту
steps_dict = {}
# Итерируемся по каждому рецепту в XML
for recipe in soup.find_all('recipe'):
    # Извлекаем id рецепта
    recipe_id = recipe.find('id').text
    # Извлекаем шаги
    steps = [step.text.strip() for step in recipe.find_all('step')]
    # Добавляем информацию о шагах в словарь
    steps_dict[recipe_id] = steps
# Записываем словарь в файл JSON
with open('steps_sample.json', 'w') as json_file:
    json.dump(steps_dict, json_file, indent=4)
````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
# Инициализируем словарь для хранения количества шагов и списков их идентификаторов
steps_count_dict = {}
# Итерируемся по каждому рецепту в XML
for recipe in soup.find_all('recipe'):
    # Извлекаем id рецепта
    recipe_id = recipe.find('id').text
    
    # Извлекаем количество шагов
    steps_count = len(recipe.find_all('step'))
    
    # Добавляем id рецепта в список соответствующего количеству шагов
    if steps_count in steps_count_dict:
        steps_count_dict[steps_count].append(recipe_id)
    else:
        steps_count_dict[steps_count] = [recipe_id]

# Записываем словарь в файл JSON
with open('steps_count_sample.json', 'w') as json_file:
    json.dump(steps_count_dict, json_file, indent=4)
````

3.3 Получите список рецептов, в этапах выполнения которых есть информация о времени (часы или минуты). Для отбора подходящих рецептов обратите внимание на атрибуты соответствующих тэгов.

````python
# Инициализируем список для хранения рецептов с информацией о времени
recipes_with_time_info = []

# Итерируемся по каждому рецепту в XML
for recipe in soup.find_all('recipe'):
    recipe_id = recipe.find('id').text
    steps = recipe.find_all('step')
    for step in steps:
        # Проверяем наличие атрибутов has_minutes и has_degrees
        if 'has_minutes' in step.attrs or 'has_degrees' in step.attrs:
            recipes_with_time_info.append(recipe_id)
            break  # Прерываем внутренний цикл, если нашли хотя бы один шаг с информацией о времени
            

# Выводим список рецептов с информацией о времени
print("Рецепты с информацией о времени:")
print(recipes_with_time_info)
````

3.4 Загрузите данные из файла `recipes_sample.csv` (__ЛР2__) в таблицу `recipes`. Для строк, которые содержат пропуски в столбце `n_steps`, заполните этот столбец на основе файла  `steps_sample.xml`. Строки, в которых столбец `n_steps` заполнен, оставьте без изменений.

````python
# Создаем список индексов строк, где пропущены значения в столбце n_steps
missing_steps_indices = recipes[recipes['n_steps'].isnull()].index

# Создаем пустой словарь для хранения количества шагов для каждого рецепта в XML
steps_count_dict = {}
# Проходимся по каждому XML-рецепту, извлекаем идентификатор и подсчитываем количество шагов
for recipe in soup.find_all('recipe'):
    recipe_id = recipe.find('id').text
    steps_count = len(recipe.find_all('step'))
    steps_count_dict[recipe_id] = steps_count  # Сохраняем количество шагов для каждого рецепта в словаре
# Заполняем пропущенные значения в столбце n_steps на основе данных из XML
for index in missing_steps_indices:  # Проходимся по индексам строк с пропущенными значениями
    recipe_id = recipes.loc[index, 'id']  # Извлекаем идентификатор рецепта из DataFrame
    if str(recipe_id) in steps_count_dict:  # Проверяем, есть ли данные о количестве шагов для данного рецепта в словаре
        recipes.at[index, 'n_steps'] = steps_count_dict[str(recipe_id)]  # Заполняем пропуск количеством шагов из словаря
# Выводим DataFrame с заполненными пропущенными значениями
recipes
````
